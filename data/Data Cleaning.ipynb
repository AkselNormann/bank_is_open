```python
team = "Golden State Warriors"
team_games = season[season.team == team]
```

```python
team_list = pd.unique(season.team)
correlations = [0 for j in range(len(team_games.columns[6:]))]

for i, col in enumerate(team_games.columns[6:]):
    tot_data = []
    tot_out = []
    
    for team in team_list:
        team_games = season[season.team == team]
        out = np.array(team_games.team_PTS + team_games.opponent_PTS)
        out = out[1:]
        tot_out.extend(out)

        data = np.array(team_games[col])[:-1]
        tot_data.extend(data)
    
    correlations[i] = np.corrcoef(tot_data,tot_out)[1][0]
        
    #break
        
correlations = np.array(correlations)
```

```python
cols = season.columns
corr = np.abs(correlations)
corr, cols = zip(*sorted(zip(corr, cols), reverse=True))
corr = np.array(corr)
cols = np.array(cols)
```

```python
cols
```

```python
season = season.drop(["team_FTAr", 'Unnamed: 0'], axis = 1)
to_remove = list(cols[corr < 0.08])
to_keep = ["date", "team", "opponent"]
for t in to_keep:
    if t in to_remove: to_remove.remove(t)
```

```python
season = season.drop(to_remove, axis = 1)
```
